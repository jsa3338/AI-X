# 1절. 조건문(if)
```
if 조건1:
    조건1이 참일 경우 수행할 명령어(들)
elif 조건2:
    조건1이 거짓이고, 조건2가 참일 경우 수행할 명령어(들)
else:
    모든 조건을 만족하지 않을때 수행할 명령어(들)
# elif, else 생략가능

```

````python
score = int(input('점수는 : '))
if score >=90:               
    print('참 잘했어요')            #반드시 명령문이 1개 이상 있어야 하며, 넘어가야 할 경우 pass를 사용
    print('다음에도 기대할게요')
elif score >= 60:
    print('분발하세요')            
    print('고생은 했어요')
else:
    print('재시험 대상')            
    print('힘을 내요')
print('화이팅!')
````

````python
# 점수(0~100)를 입력받아 학점을 출력하시오(A, B, C, D, F, 유효하지 않은 점수)

score = int(input('점수는 : '))
if (score > 100) | (score<0):
    grade = '유효하지 않은 점수'
elif score >=90 :               
    grade = 'A'           
elif score >= 80:
    grade = 'B'          
elif score >= 70:
    grade = 'C'          
elif score >= 60:
    grade = 'D'           
else:
    grade = 'F'
print(f'입력하신 점수는 {score}이고, {grade} 등급입니다.')   
````

## 2.1 for-in문
```
for 임시변수 in 나열가능한자료:
    반복할명령어1
    반복할명령어2
    if 조건:
        break
else:
    나열가능한자료가 다 소모되어 for블록을 다 수행한 후 **최종 한 번 수행할 명령어**
    for문 안에 break를 수행할 경우 else블록은 수행되지 않음
```

````python
for i in range(1, 10, 2):
    print(i, end='\t')
else:
    print('\nDONE')
print('END')
````

````python
for i in range(1, 10, 2):
    if i ==5:
        break
    print(i, end='\t')      
else:
    print('\nDONE')
print('END')
````

````python
for i in range(1, 10, 2):
    if i ==5:
        continue
    print(i, end='\t')      
else:
    print('\nDONE')
print('END')
````

## 2.2 while문
```
while 조건:
    조건이 참일 경우 반복할 문장들

    조건이 거짓일 경우 한번 실행할 문장(else구문은 선택사항)
```

````python
# ex2. 선생님 풀이 - 리스트 이용
lotto = []
cnt = 0
while len(lotto) < 6:
    cnt +=1
    temp = random.randint(1,45)
    if temp not in lotto:
        lotto.append(temp)
# 정렬하기  lotto.sort()
print(f'난수 {cnt}번 뽑아서 로또는 {sorted(lotto)}')
````

````python
# 정렬하기  lotto.sort() 구현하기

for i in range(0, len(lotto)-1):
    for j in range(i, len(lotto)):
        if lotto[i] > lotto[j]:
            lotto[i], lotto[j] = lotto[j], lotto[i] 
    print(f'{i}번째 까지 : {lotto[:i+1]} / {lotto}')
            
````

````python
# 10. (3단계) 1부터 100 사이의 정수를 맞추는 프로그램을 구현하시오
'''
3단계. 숫자를 맞출 수 있는 기회를 5회로 제한하고 5회 이내에 맞추지 않으면 실패
메세지를 출력합니다
'''
min_=1
max_=100
cnt_=5
random_num = random.randint(min_,max_)
print('컴퓨터가 발생한 임의의 수 :', random_num)

while cnt_>=1:
    
    input_num = int(input(f'{min_}~{max_} 사이의 숫자를 맞춰보세요(남은기회는 {cnt_}번) : '))
    if (min_ > input_num) | (max_ < input_num):
        print(f'수의 범위를 벗어났습니다')
    elif (random_num > input_num):
        print('UP ↑')
        min_ = input_num+1
        cnt_ -=1
    elif (random_num < input_num):
        print('DOWN ↓')
        max_ = input_num-1
        cnt_ -=1
    else:
        print('축하합니다. 맞추셨습니다')
        break
else:
    print('기회를 모두 사용하셨습니다. 실패하셨습니다')
````

````python
# 10. 선생님 풀이
#(3단계) 1부터 100 사이의 정수를 맞추는 프로그램을 구현하시오 <- 횟수가 정해져있으므로 for 구문으로 가는게 낫다
'''
3단계. 숫자를 맞출 수 있는 기회를 5회로 제한하고 5회 이내에 맞추지 않으면 실패
메세지를 출력합니다
'''
min_, max_=1, 100
random_num = random.randint(min_,max_)
print('컴퓨터가 발생한 임의의 수 :', random_num)

for i in range(5, 0, -1):
    input_num = int(input(f'{min_}~{max_} 사이의 숫자를 맞춰보세요(남은기회는 {i}번) : '))
    if (min_ > input_num) | (max_ < input_num):
        print(f'수의 범위를 벗어났습니다')
    elif (random_num > input_num):
        print('UP ↑')
        min_ = input_num+1
    elif (random_num < input_num):
        print('DOWN ↓')
        max_ = input_num-1
    else:
        print('축하합니다. 맞추셨습니다')
        break
else:
    print('기회를 모두 사용하셨습니다. 실패하셨습니다')
````
