```python
import numpy as np

def start_circle_area_game():
    print("ğŸ® ì›ì˜ ë„“ì´ ë§íˆê¸° ê²Œì„ ì‹œì‘!")
    while True:
        r = np.random.randint(1, 21)  # 1ë¶€í„° 20ê¹Œì§€ ëœë¤ ë°˜ì§€ë¦„
        correct_area = round(np.pi * (r ** 2), 2)

        print(f"\nğŸŒ€ ë°˜ì§€ë¦„ r = {r}")
        guess = input("ë„“ì´ëŠ” ì–¼ë§ˆì¼ê¹Œìš”? (ì†Œìˆ˜ì  2ìë¦¬ê¹Œì§€): ")

        try:
            guess = float(guess)
            diff = abs(guess - correct_area)
            if diff < 0.5:
                print("âœ… ì •ë‹µ! ê±°ì˜ ì™„ë²½í•´ìš”!")
            else:
                print(f"âŒ ì•„ì‰½! ì •ë‹µì€ {correct_area}ì˜€ì–´ìš”. (ì°¨ì´: {diff:.2f})")
        except:
            print("âš ï¸ ìˆ«ìë¡œ ì…ë ¥í•´ì¤˜!")

        again = input("ë‹¤ì‹œ í• ë˜ìš”? (y/n): ")
        if again.lower() != "y":
            print("ğŸ‘‹ ê²Œì„ ë! ìˆ˜ê³ í–ˆì–´~")
            break

# ì‹¤í–‰
start_circle_area_game()
```

```python
import random

answer = random.randint(1, 100)
tries = 0

print("ğŸ² ìˆ«ì ë§ì¶”ê¸° ê²Œì„ ì‹œì‘! (1~100 ì‚¬ì´)")
print("íŒíŠ¸: ìˆ«ìë¥¼ ì…ë ¥í•˜ë©´ 'ì—…' ë˜ëŠ” 'ë‹¤ìš´'ìœ¼ë¡œ ì•Œë ¤ì¤˜ìš”!")

while True:
    user = input("ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš”: ")

    # ìˆ«ìì¸ì§€ í™•ì¸
    if not user.isdigit():
        print("â— ìˆ«ìë§Œ ì…ë ¥í•´ì£¼ì„¸ìš”!")
        continue

    guess = int(user)
    tries += 1

    if guess < answer:
        print("ğŸ”¼ ì—…! ë” í° ìˆ˜ë¥¼ ì…ë ¥í•´ë³´ì„¸ìš”.")
    elif guess > answer:
        print("ğŸ”½ ë‹¤ìš´! ë” ì‘ì€ ìˆ˜ë¥¼ ì…ë ¥í•´ë³´ì„¸ìš”.")
    else:
        print(f"ğŸ‰ ì •ë‹µì…ë‹ˆë‹¤! {tries}ë²ˆ ë§Œì— ë§ì·„ì–´ìš”!")
        break
```
