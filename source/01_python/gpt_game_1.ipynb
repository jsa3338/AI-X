```python
import numpy as np

def start_circle_area_game():
    print("🎮 원의 넓이 맞히기 게임 시작!")
    while True:
        r = np.random.randint(1, 21)  # 1부터 20까지 랜덤 반지름
        correct_area = round(np.pi * (r ** 2), 2)

        print(f"\n🌀 반지름 r = {r}")
        guess = input("넓이는 얼마일까요? (소수점 2자리까지): ")

        try:
            guess = float(guess)
            diff = abs(guess - correct_area)
            if diff < 0.5:
                print("✅ 정답! 거의 완벽해요!")
            else:
                print(f"❌ 아쉽! 정답은 {correct_area}였어요. (차이: {diff:.2f})")
        except:
            print("⚠️ 숫자로 입력해줘!")

        again = input("다시 할래요? (y/n): ")
        if again.lower() != "y":
            print("👋 게임 끝! 수고했어~")
            break

# 실행
start_circle_area_game()
```

```python
import random

answer = random.randint(1, 100)
tries = 0

print("🎲 숫자 맞추기 게임 시작! (1~100 사이)")
print("힌트: 숫자를 입력하면 '업' 또는 '다운'으로 알려줘요!")

while True:
    user = input("숫자를 입력하세요: ")

    # 숫자인지 확인
    if not user.isdigit():
        print("❗ 숫자만 입력해주세요!")
        continue

    guess = int(user)
    tries += 1

    if guess < answer:
        print("🔼 업! 더 큰 수를 입력해보세요.")
    elif guess > answer:
        print("🔽 다운! 더 작은 수를 입력해보세요.")
    else:
        print(f"🎉 정답입니다! {tries}번 만에 맞췄어요!")
        break
```

```python
import random

# 플레이어 상태
player = {
    "이름": "용사",
    "체력": 100,
    "공격력": 20,
    "회복약": 3,
    "레벨": 1,
    "경험치": 0
}

# ────────────────────────
# 1. 몬스터 생성 함수
def spawn_monster():
    몬스터 = {
        "이름": random.choice(["고블린", "슬라임", "해골 전사"]),
        "체력": random.randint(30, 60),
        "공격력": random.randint(5, 15)
    }
    return 몬스터

# ────────────────────────
# 2. 전투 함수
def fight(몬스터):
    print(f"\n⚔️ {몬스터['이름']}이(가) 나타났다!")

    while 몬스터["체력"] > 0 and player["체력"] > 0:
        action = input("무엇을 할까? (1) 공격 (2) 회복 (3) 도망 : ")

        if action == "1":
            damage = random.randint(10, player["공격력"])
            몬스터["체력"] -= damage
            print(f"💥 {몬스터['이름']}에게 {damage}의 피해를 줬다!")

        elif action == "2":
            if player["회복약"] > 0:
                player["체력"] += 30
                player["회복약"] -= 1
                print("🧪 체력을 30 회복했다! 남은 회복약:", player["회복약"])
            else:
                print("❌ 회복약이 없다!")

        elif action == "3":
            print("🏃 도망쳤다!")
            return  # 함수 종료

        else:
            print("잘못된 입력이야!")
            continue

        # 몬스터 반격
        if 몬스터["체력"] > 0:
            반격 = random.randint(5, 몬스터["공격력"])
            player["체력"] -= 반격
            print(f"😖 {몬스터['이름']}의 반격! {반격}의 피해를 입었다.")

    # 전투 종료 후 처리
    if player["체력"] <= 0:
        print("💀 당신은 쓰러졌다... 게임 오버")
        exit()

    else:
        print(f"🎉 {몬스터['이름']}을(를) 처치했다!")
        gained_exp = random.randint(10, 25)
        player["경험치"] += gained_exp
        print(f"✨ 경험치 {gained_exp} 획득!")

        # 레벨업
        if player["경험치"] >= 50:
            player["레벨"] += 1
            player["공격력"] += 5
            player["체력"] = 100
            player["경험치"] = 0
            print(f"⬆️ 레벨 업! 현재 레벨: {player['레벨']}")

# ────────────────────────
# 3. 게임 루프
print("🧙‍♂️ 간단한 파이썬 RPG에 오신 것을 환영합니다!")

while True:
    input("\n👉 엔터를 눌러 모험을 계속하세요...")
    monster = spawn_monster()
    fight(monster)
```
