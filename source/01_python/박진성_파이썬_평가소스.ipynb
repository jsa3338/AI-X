```python
# 문제 2
df_last = pd.read_csv(r'C:\ai_x\download\shareData\부동산_250213\주택도시보증공사_전국 신규 민간아파트 분양가격 동향_20241231.csv',
                     encoding='cp949')
```

```python
# 전용면적 열 추가
df_last['전용면적'] = df_last['규모구분'].str.replace("전용면적","")\
                    .str.replace('제곱미터', '㎡')\
                    .str.replace('초과', '~')\
                    .str.replace('이하', '')
```

```python
# 분양가격(제곱미터) 결측치 개수, 전체 행수
df_last.isna().sum(), df_last['분양가격(제곱미터)'].shape
```

```python
df_last['분양가격(제곱미터)'].astype('float') # ValueError: could not convert string to float: '  '
```

```python
# 평당분양가격 추가를 위한 분양가격(제곱미터) 처리 가공
def prepros(row):
    row = row.copy()
    price = row['분양가격(제곱미터)']
    if pd.isna(price):
        row['임시분양가격'] = np.nan
    elif price.strip() == '':
        row['임시분양가격'] = np.nan
    elif price.find(',') != -1:
        row['임시분양가격'] = int(price.replace(',', ''))
    else: 
        row['임시분양가격'] = price
    return row
```

```python
df_last = df_last.apply(prepros, axis=1)
```

```python
df_last['임시분양가격']= df_last['임시분양가격'].astype('float')
```

```python
df_last['평당분양가격'] = df_last['임시분양가격']*3.3
```

```python
df_last = df_last.drop(columns=['규모구분', '분양가격(제곱미터)', '임시분양가격'], axis=1)
df_last.head()
```

```python
df_last = df_last[['지역명', '전용면적', '연도','월', '평당분양가격']]
```

```python
df_first = df_first.melt(id_vars='지역',
             var_name='연월',
             value_name='평당분양가격')
df_first
```

```python
def sep_year(row):
    row = row.copy()
    date = row['연월']
    row['연도'] = date.split('년')[0]
    return row

def sep_month(row):
    row = row.copy()
    date = row['연월']
    row['월'] = date.split('년')[1].replace('월','')
    return row

df_first = df_first.apply(sep_year, axis=1)
df_first = df_first.apply(sep_month, axis=1)
df_first = df_first.drop('연월', axis=1)
df_first
```

```python
df_first = df_first.rename(columns={'지역' : '지역명'})
df_first = df_first[['지역명', '연도', '월', '평당분양가격']]
```
